```python
from sklearn.datasets import load_diabetes

diabetes_dataset = load_diabetes()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabetes_dataset.data, diabetes_dataset.target, random_state =0)
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor

for index in range(1, 100):
  knn_model = KNeighborsRegressor(n_neighbors = index)

  knn_model.fit(X_train, y_train)

  print("Number of neighbors: ", index)

  print (knn_model.score(X_test, y_test))
```

```python
knn_model = KNeighborsRegressor(n_neighbors = 30)

knn_model.fit(X_train, y_train)

print (knn_model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(X_train, y_train)
forest_model.score(X_test, y_test)
```

```python
from sklearn.neural_network import MLPRegressor

neural_network = MLPRegressor()

neural_network.fit(X_train, y_train)
neural_network.score(X_test, y_test)
```
